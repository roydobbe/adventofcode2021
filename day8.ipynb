```python
import numpy as np
from collections import Counter

with open('data/day8.txt', 'r') as line:
    x = line.read().splitlines()

input = [item for sublist in [digit.split(' | ')[1].split(' ') for digit in x] for item in sublist]

known_digits_lengths = [2,3,4,7]

digit_dict = Counter([len(item) for item in input])

digit_dict_known = {known_digit: digit_dict[known_digit] for known_digit in known_digits_lengths}

sum(digit_dict_known.values())
```

```python
input = [[digit.split(' | ')[0].split(' '), digit.split(' | ')[1].split(' ')] for digit in x]

digit_list = []

for y in input:
    digit_dict = {}
    one = [x for x in [s for s in y[0] if len(s) == 2][0]]
    four = [x for x in [s for s in y[0] if len(s) == 4][0]]
    seven = [x for x in [s for s in y[0] if len(s) == 3][0]]
    eight = [x for x in [s for s in y[0] if len(s) == 7][0]]

    for item in y[0]:
        counter_eight = sum([item.count(i) for i in eight])
        counter_one = sum([item.count(i) for i in one])
        counter_seven = sum([item.count(i) for i in seven])
        counter_four = sum([item.count(i) for i in four])

        if len(item) == 6:
            if counter_one == 2 and counter_four == 3:
                zero = [x for x in item]
            elif counter_one == 2:
                nine = [x for x in item]
            else:
                six = [x for x in item]
        elif len(item) == 5:
            if counter_one == 2 and counter_four == 3 and counter_seven == 3:
                three = [x for x in item]
            elif counter_one == 1 and counter_four == 3 and counter_seven == 2:
                five = [x for x in item]
            else:
                two = [x for x in item]

    digit_dict[''.join(sorted(zero))] = '0'
    digit_dict[''.join(sorted(one))] = '1'
    digit_dict[''.join(sorted(two))] = '2'
    digit_dict[''.join(sorted(three))] = '3'
    digit_dict[''.join(sorted(four))] = '4'
    digit_dict[''.join(sorted(five))] = '5'
    digit_dict[''.join(sorted(six))] = '6'
    digit_dict[''.join(sorted(seven))] = '7'
    digit_dict[''.join(sorted(eight))] = '8'
    digit_dict[''.join(sorted(nine))] = '9'

    total_digit = ''

    for digit in y[1]:
        total_digit = total_digit + digit_dict[''.join(sorted(digit))]
    
    digit_list.append(int(total_digit))

sum(digit_list)
```
