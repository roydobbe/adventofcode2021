```python
import pandas as pd
```

```python
input = pd.read_csv('data/day3.csv', dtype={'input': object})
```

```python
gamma = int(''.join(input['input'].apply(lambda x: pd.Series(list(x))).astype(int).agg({'sum', 'count'}).T.apply(lambda x: '1' if x['sum'] > x['count'] / 2 else '0', axis=1).to_list()), 2)
epsilon = int(''.join(input['input'].apply(lambda x: pd.Series(list(x))).astype(int).agg({'sum', 'count'}).T.apply(lambda x: '1' if x['sum'] < x['count'] / 2 else '0', axis=1).to_list()), 2)

gamma * epsilon
```

```python
output_ox = input.copy()
output_co = input.copy()

for i in range(0, len(input['input'].loc[0])):
    filter = output_ox['input'].apply(lambda x: pd.Series(list(x))).astype(int).agg({'sum', 'count'}).T.apply(lambda x: '1' if x['sum'] >= x['count'] / 2 else '0', axis=1)[i]
    output_ox = output_ox[output_ox['input'].str[i] == filter]
    if len(output_ox.index) == 1:
        ox = int(output_ox['input'].iloc[0], 2)
        break


for x in range(0, len(input['input'].loc[0])):
    filter = output_co['input'].apply(lambda x: pd.Series(list(x))).astype(int).agg({'sum', 'count'}).T.apply(lambda x: '0' if x['sum'] >= x['count'] / 2 else '1', axis=1)[x]
    output_co = output_co[output_co['input'].str[x] == filter]
    if len(output_co.index) == 1:
        co = int(output_co['input'].iloc[0], 2)
        break

co * ox
```
