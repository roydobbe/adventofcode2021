```python
import numpy as np

with open('data/day5.txt', 'r') as line:
    x = line.readlines()

lines = [i.replace(' -> ',',').replace('\n', '') for i in x]

input = [[[int(line.split(',')[0]), int(line.split(',')[1])], [int(line.split(',')[2]), int(line.split(',')[3])]] for line in lines]

max_range = max([int(value) for value in[max(line.split(',')) for line in lines]])

result_array = np.zeros([max_range + 1, max_range + 1])


for item in input:
    if item[0][0] == item[1][0]:
        for x in range(min([item[0][1], item[1][1]]), max([item[0][1], item[1][1]]) + 1):
            result_array[item[0][0], x] += 1
    elif item[0][1] == item[1][1]:
        for x in range(min([item[0][0], item[1][0]]), max([item[0][0], item[1][0]]) + 1):
            result_array[x, item[0][1]] += 1
    else:
        pass

np.count_nonzero(result_array >= 2)
```

```python
with open('data/day5.txt', 'r') as line:
    x = line.readlines()

lines = [i.replace(' -> ',',').replace('\n', '') for i in x]

input = [[[int(line.split(',')[0]), int(line.split(',')[1])], [int(line.split(',')[2]), int(line.split(',')[3])]] for line in lines]

max_range = max([int(value) for value in[max(line.split(',')) for line in lines]])

result_array = np.zeros([max_range + 1, max_range + 1])


for item in input:
    x1 = item[0][0]
    y1 = item[0][1]
    x2 = item[1][0]
    y2 = item[1][1]

    if x1 == x2:
        for x in range(min([y1, y2]), max([y1, y2]) + 1):
            result_array[x1, x] += 1
    elif y1 == y2:
        for x in range(min([x1, x2]), max([x1, x2]) + 1):
            result_array[x, y1] += 1
    elif x1 - x2 == y1 - y2:
        x_axis = min([x1, x2])
        y_axis = min([y1, y2])
        for x in range(x_axis, max([x1, x2])+1):
            if x_axis <= max([x1, x2]):
                result_array[x_axis, y_axis] += 1
                x_axis += 1
                y_axis += 1
    elif x2 - x1 == y1 - y2:
        x_axis = min([x1, x2])
        y_axis = max([y1, y2])
        for x in range(x_axis, max([x1, x2])+1):
            if x_axis <= max([x1, x2]):
                result_array[x_axis, y_axis] += 1
                x_axis += 1
                y_axis -= 1
    else:
        pass

np.count_nonzero(result_array >= 2)
```
