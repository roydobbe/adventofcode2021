```python
import numpy as np
import re

lower = '^[a-z]{2}$'

with open('data/day12.txt', 'r') as line:
    x = line.read().splitlines()


locations = set([item for sublist in [item.split('-') for item in x] for item in sublist])

paths = [item.split('-') for item in x]
```

```python
smoll = [x for x in locations if re.search(lower, x)]

```

```python
location_path_dict = {}

for location in locations:
    location_paths = []
    for path in paths:
        if location in path:
            pathway = [x for x in path if location not in x][0]
            location_paths.append(pathway)
    location_path_dict[location] = location_paths

for key, value in location_path_dict.items():
    location_path_dict[key] = [x for x in value if 'start' not in x]

location_path_dict
```

```python
import random

locations

path_list = []


for retry in range(10000000):
    
    loop = 0
    location_path_dict_copy = location_path_dict.copy()
    next = 'start'
    path = 'start'

    while loop == 0:
        try:
            next = location_path_dict_copy[next][random.randrange(len(location_path_dict_copy[next]))]
            path += f'-{next}'
            if next in smoll:
                for key, value in location_path_dict_copy.items():
                    location_path_dict_copy[key] = [x for x in value if next not in x]

            if next == 'end':
                path_list.append(path)
                loop = 1
        except:
            loop = 1
    else:
        loop = 1


len(set(path_list))
```

```python
smoll_dict = {}

for sm in smoll:
    smoll_dict[sm] = 1

path_list = []

for retry in range(10000000):
    
    loop = 0
    location_path_dict_copy = location_path_dict.copy()
    next = 'start'
    path = 'start'

    while loop == 0:
        try:
            next = location_path_dict_copy[next][random.randrange(len(location_path_dict_copy[next]))]
            path += f'-{next}'
            if next in smoll:
                if smoll_dict[next] == 1:
                    smoll_dict[next] -= 1
                else:
                    for key, value in location_path_dict_copy.items():
                        location_path_dict_copy[key] = [x for x in value if next not in x]

            if next == 'end':
                path_list.append(path)
                loop = 1
        except:
            loop = 1
    else:
        loop = 1


len(set(path_list))
```
