```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/day2.csv')
df[['direction', 'amount']] = df['movement'].str.split(" ",expand=True,)
df['amount'] = df['amount'].astype(int)

df.groupby('direction').sum()
```

```python
df = df[['direction', 'amount']]
df['aim_increase'] =  np.where(df['direction'] == 'down', 1 * df['amount'], np.where(df['direction'] == 'up', -1 * df['amount'],0))
df['aim'] = df.aim_increase.cumsum()
df['depth'] = np.where(df['direction'] == 'forward', df['aim'] * df['amount'], 0)
df.groupby('direction').sum()
```
