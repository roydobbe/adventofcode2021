```python
numbers = []

for y in range(len(field)):
        for x in range(len(field[y])):
            if (x == 0) & (y == 0):
                right = field[y][x+1]
                under = field[y+1][x]
                if (field[y][x] < right) & (field[y][x] < under):
                    numbers.append(field[y][x])
            elif (x == 0) & (y == len(field) - 1):
                right = field[y][x+1]
                upper = field[y-1][x]
                if (field[y][x] < right) & (field[y][x] < upper):
                    numbers.append(field[y][x])
            elif (x == len(field[0]) - 1) & (y == 0):
                left = field[y][x-1]
                under = field[y+1][x]
                if (field[y][x] < under) & (field[y][x] < left):
                    numbers.append(field[y][x])
            elif (x == len(field[0]) - 1) & (y == len(field) - 1):
                left = field[y][x-1]
                upper = field[y-1][x]
                if (field[y][x] < upper) & (field[y][x] < left):
                    numbers.append(field[y][x])
            elif (x == 0) & (y != 0):
                upper = field[y-1][x]
                right = field[y][x+1]
                under = field[y+1][x]
                if (field[y][x] < upper) & (field[y][x] < under) & (field[y][x] < right):
                    numbers.append(field[y][x])

            elif (x != 0) & (y == 0):
                left = field[y][x-1]
                right = field[y][x+1]
                under = field[y+1][x]
                if (field[y][x] < right) & (field[y][x] < under) & (field[y][x] < left):
                    numbers.append(field[y][x])
            elif (x != len(field[0]) - 1) & (y == len(field) - 1):
                left = field[y][x-1]
                upper = field[y-1][x]
                right = field[y][x+1]
                if (field[y][x] < left) & (field[y][x] < right) & (field[y][x] < upper):
                    numbers.append(field[y][x])
            elif (x == len(field[0]) - 1) & (y != len(field) - 1):
                left = field[y][x-1]
                upper = field[y-1][x]
                under = field[y+1][x]
                if (field[y][x] < left) & (field[y][x] < under) & (field[y][x] < upper):
                    numbers.append(field[y][x])
            else:
                left = field[y][x-1]
                upper = field[y-1][x]
                right = field[y][x+1]
                under = field[y+1][x]
                if (field[y][x] < upper) & (field[y][x] < left) & (field[y][x] < right) & (field[y][x] < under):
                    numbers.append(field[y][x])

numbers_added = [int(x) + 1 for x in numbers]

sum(numbers_added)
```
