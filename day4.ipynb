```python
import numpy as np
from io import StringIO


#open text file in read mode
text_file = open("data/day4.txt", "r")
 
#read whole file to a string
data = text_file.read()
 
#close file
text_file.close()

def divide_chunks(l, n):
      
    # looping till length l
    for i in range(0, len(l), n): 
        yield l[i:i + n]

bingo = np.genfromtxt(StringIO(data), dtype=int)
  
# How many elements each
# list should have
n = 5
  
bingo_divided = list(divide_chunks(bingo, 5))
# print (x)

```

```python
for number in bingo_list:
    bingo_divided[44] = np.where(bingo_divided[44] == number, 0, bingo_divided[44])
    check_bingo = list(bingo_divided[44].sum(axis=1)) + list(bingo_divided[44].sum(axis=0))

    if 0.in(check_bingo):
        print('Bingo!')
        print(bingo_divided[44].sum() * number)
        break
```

```python
for number in bingo_list:
    for bingo_card in range(0,len(bingo_divided)):
        bingo_divided[bingo_card] = np.where(bingo_divided[bingo_card] == number, 0, bingo_divided[bingo_card])
        check_bingo = list(bingo_divided[bingo_card].sum(axis=1)) + list(bingo_divided[bingo_card].sum(axis=0))

        if 0.in(check_bingo):
            print('Bingo!')
            print(bingo_divided[bingo_card].sum() * number)
            break
```

```python
bingo_dict = {}
n=0


for number in bingo_list:
    for bingo_card in range(0,len(bingo_divided)):
        bingo_divided[bingo_card] = np.where(bingo_divided[bingo_card] == number, 0, bingo_divided[bingo_card])
        check_bingo = list(bingo_divided[bingo_card].sum(axis=1)) + list(bingo_divided[bingo_card].sum(axis=0))

        if 0.in(check_bingo):
            n += 1
            if bingo_card in bingo_dict:
                bingo_dict[bingo_card].append(n)
            else:
                bingo_dict[bingo_card] = [n]
```

```python
first_bingo = {}

for key, value in bingo_dict.items():
    first_bingo[key] = value[0]
    # print(value)
```
