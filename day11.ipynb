```python
import numpy as np

with open('data/day11.txt', 'r') as line:
    x = line.read().splitlines()

field = np.array([list(map(int, octopus)) for octopus in x])

steps = 100
```

```python
flash = 0
step_total = 0


for step in range(1000):
    step_total += 1
    list_9 = []
    list_explosion = []

    field += 1

    nine_checker = np.where(field > 9)

    for item in range(len(nine_checker[0])):
        list_9.append([nine_checker[0][item], nine_checker[1][item]])
    while len(list_9) > 0:
        y = list_9[0][0]
        x = list_9[0][1]
        if (y == 0) & (x == 0):
            field[y + 1][ x + 1] += 1
            field[y + 1][ x] += 1
            field[y][ x + 1] += 1
        elif (y == 9) & (x == 0):
            field[y - 1][ x] += 1
            field[y][ x + 1] += 1
            field[y - 1][ x + 1] += 1
        elif (y == 0) & (x ==9):
            field[y][ x -1] += 1
            field[y + 1][ x] += 1
            field[y + 1][ x - 1] += 1
        elif (y == 9) & (x ==9):
            field[y - 1][ x -1] += 1 
            field[y - 1][ x] += 1
            field[y][ x -1] += 1
        elif (y == 0) & (x != 0):
            field[y][ x -1] += 1
            field[y + 1][ x + 1] += 1
            field[y + 1][ x] += 1
            field[y][ x + 1] += 1
            field[y + 1][ x - 1] += 1
        elif (y != 0) & (x == 0):
            field[y - 1][ x] += 1
            field[y + 1][ x + 1] += 1
            field[y + 1][ x] += 1
            field[y][ x + 1] += 1
            field[y - 1][ x + 1] += 1
        elif (y != 9) & (x ==9):
            field[y - 1][ x -1] += 1 
            field[y - 1][ x] += 1
            field[y][ x -1] += 1
            field[y + 1][ x] += 1
            field[y + 1][ x - 1] += 1
        elif (y == 9)  & (x != 9):
            field[y - 1][ x -1] += 1 
            field[y - 1][ x] += 1
            field[y][ x -1] += 1
            field[y][ x + 1] += 1
            field[y - 1][ x + 1] += 1
        else:
            field[y - 1][ x -1] += 1 
            field[y - 1][ x] += 1
            field[y][x -1] += 1
            field[y + 1][ x + 1] += 1
            field[y + 1][ x] += 1
            field[y][ x + 1] += 1
            field[y - 1][ x + 1] += 1
            field[y + 1][ x - 1] += 1

        nine_checker = np.where(field > 9)
        list_explosion.append(list_9[0])

        for item_2 in range(len(nine_checker[0])):
            list_9.append([nine_checker[0][item_2], nine_checker[1][item_2]])

        flash += 1

        list_9 = [flash for flash in list_9 if flash not in list_explosion]   

    else:
        field = np.where(field > 9, 0, field)
        if np.all(field == field[0]):
            print(f'synched: {step + 1}')
            break
        if step == 99:
            print(f'flashes: {flash}')
```
