```python
import numpy as np
import re

regexp = re.compile('\[\]|\{\}|\(\)|\<\>')
regexp_2 = re.compile('\]|\}|\)|\>')
regexp_3 = re.compile('\[|\{|\(|\<')

with open('data/day10.txt', 'r') as line:
    x = line.read().splitlines()

score_dict = {')': 3, ']': 57, '}' : 1197, '>': 25137, '(': 1, '[': 2 , '{' : 3, '<' : 4}
```

```python
syntax_score = 0
syntax_score_list = []

for item in x:
    while bool(re.search(regexp, item)):
        item = re.sub(regexp, '', item)
    else:
        if bool(re.search(regexp_2, item)):
            item = re.sub(regexp_3, '', item)
            syntax_score += score_dict[item[0]]
        else:
            score = 0
            for bracket in item[::-1]:
                score = score * 5
                score += score_dict[bracket]
            syntax_score_list.append(score)

print(syntax_score)
print(np.median(syntax_score_list))
```
