```python
import pandas as pd
import numpy as np
```

```python
depth = pd.read_csv('data/day1.csv')

depth['previous_depth'] = depth['depth'].shift(1)

depth['status'] = np.where(
    depth['previous_depth'].astype(float) >  depth['depth'].astype(float), 'decreased',
    np.where(
        depth['previous_depth'].astype(float) < depth['depth'].astype(float), 'increased',
    np.where(
        depth['previous_depth'].astype(float) == depth['depth'].astype(float), 'stayed_the_same', 'none'
    )))

depth.status.value_counts()
```

```python
depth = pd.read_csv('data/day1.csv')

depth['rolling'] = depth.rolling(window=3).sum()

depth['previous_depth'] = depth['rolling'].shift(1)

depth['status'] = np.where(
    depth['previous_depth'].astype(float) >  depth['rolling'].astype(float), 'decreased',
    np.where(
        depth['previous_depth'].astype(float) < depth['rolling'].astype(float), 'increased',
    np.where(
        depth['previous_depth'].astype(float) == depth['rolling'].astype(float), 'stayed_the_same', 'none'
    )))

depth.status.value_counts()
```
